```python
import networkx as nx
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
# Load the New York City graph from OSM
place_name = "New York City, New York, USA"
G = ox.graph_from_place(place_name, network_type="drive")

# Choose random start and end nodes
start_node = list(G.nodes())[0]
end_node = list(G.nodes())[9020]
```

```python
start_node2 = list(G.nodes())[25]
end_node2 = list(G.nodes())[1020]
```

```python
optimal_path = nx.shortest_path(G, source=start_node, target=end_node, weight='length')
optimal_path2=nx.shortest_path(G, source=start_node2, target=end_node2, weight='length')
```

```python
fig, ax = ox.plot_graph_routes(G, routes=[optimal_path,optimal_path2], route_colors=['g','r'], route_linewidth=3, node_size=0)

plt.show()
```

```python
fig, ax = ox.plot_graph_route(G, optimal_path, route_color='r', route_linewidth=3, node_size=0)

plt.show()
```
