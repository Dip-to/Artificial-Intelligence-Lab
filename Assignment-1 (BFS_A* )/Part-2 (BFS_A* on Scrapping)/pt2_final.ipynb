```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from collections import deque

def is_subset(subset, string):
    # Convert both the subset and the string to lowercase for case-insensitive matching
    subset = subset.lower()
    string = string.lower()

    # Check if the subset is present in the string
    return subset in string
def fileClear(st):
    if(st=="DFS"):
        file = open("DFS_links.txt", "w")
        file.write("")
        file.close()
        file = open("DFS_text.txt", "w")
        file.write("")
        file.close()
        
    if(st=="BFS"):
        file = open("BFS_links.txt", "w")
        file.write("")
        file.close()
        file = open("BFS_text.txt", "w")
        file.write("")
        file.close()
        
        

def fileAdd(st,data):
    if(st=="DFS"):
        file = open("DFS_links.txt", "a")
        file.write(data + "\n")
        file.close()
        
    if(st=="BFS"):
        file = open("BFS_links.txt", "a")
        file.write(data + "\n")
        file.close()
    
    

def fileAddText(st,data):
    if(st=="DFS"):
        file = open("DFS_text.txt", "a")
        file.write(data + "\n")
        file.close()
        
    if(st=="BFS"):
        file = open("BFS_text.txt", "a")
        file.write(data + "\n")
        file.close()

def scrape_pTag_text(driver):
    pTag = driver.find_elements(By.TAG_NAME, "p")
    text = ""
    for tag in pTag:
        text += tag.text
    return text

def get_all_links_with_selenium(st,url):
    try:
        driver = webdriver.Chrome()
        driver.set_page_load_timeout(10)
        driver.get(url)
        text = scrape_pTag_text(driver)
        fileAddText(st,text)
        elements = driver.find_elements(By.XPATH, "//a[@href]")
        urls = []
        for element in elements:
            href = element.get_attribute("href")
            if href:
                if not ("google" in href or "facebook" in href or "youtube" in href or "twitter" in href or "instagram" in href or "linkedin" in href):
                    if href not in urls:
                        urls.append(href)
                        fileAdd(st,href)
        driver.quit()
        return urls
    except Exception as e:
        print(e)
        return []

def googleSearch(search):
    get_all_links_with_selenium('https://www.google.com/search?q=' + search)

def main():
    googleSearch("Dhaka")

class link:
    def __init__(self, link, parent, visited, depth):
        self.link = link
        self.parent = parent
        self.visited = visited
        self.depth = depth

class Search:
    def __init__(self):
        self.linkset = set()
        self.Graph = {}

    def DFS(self, graph, node,inp, visited = None, distance=0):
        if not node.startswith("http"):
            return
        if distance >1:
            return
        if not is_subset(inp,node):
            return
        distance += 1
        if visited is None:
            visited = set()
        if node not in visited:
            visited.add(node)
            urls = get_all_links_with_selenium("DFS",node)
            print(node, distance)
            if urls:
                print(urls[0])
            graph[node] = urls
            for neighbour in graph[node]:
                self.DFS(graph, neighbour,inp, visited, distance)
                
                
    def BFS(self, graph, start_node, inp):
        if not start_node.startswith("http"):
            return
        
        visited = set()
        queue = deque([(start_node, 0)])  # Queue of tuples (node, distance)
        
        while queue:
            node, distance = queue.popleft()
            
            if node in visited or distance > 1:
                continue
            
            if not is_subset(inp, node):
                continue
            
            visited.add(node)
            urls = get_all_links_with_selenium("BFS",node)
            print(node, distance)
            if urls:
                print(urls[0])
            graph[node] = urls
            
            for neighbour in urls:
                if neighbour not in visited:
                    queue.append((neighbour, distance + 1))

my_graph = Search()

inp=input('What would you like to search on Google?\n')
start_node = f"https://www.google.com/search?q={str(inp)}"
print("DFS starting from node", start_node)
fileClear("DFS")
my_graph.Graph = {start_node: []}
visited = None
my_graph.DFS(my_graph.Graph, start_node,inp=str(inp) )
print('done')


```
